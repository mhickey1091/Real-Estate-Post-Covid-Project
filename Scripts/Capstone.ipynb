# Bringing SQL data back into python after organizing it.

```python
from sqlalchemy import create_engine, text
```

```python
database_name = 'realpop'    # Fill this in with your lahman database name

connection_string = f"postgresql://postgres:postgres@localhost:5432/{database_name}"
```

```python
engine = create_engine(connection_string)
```

```python
def sql(query):
    with engine.connect() as connection :
        return pd.read_sql(text(query), con = connection)
```

```python
sql_table = '''SELECT year, 
		state, 
		sum(homes_sold) as homes_sold
FROM realpop
GROUP BY 1,2
ORDER BY year; '''
```

```python
homes_sold_years = sql(sql_table)
homes_sold_years
```

# Adding two new columns so I can calculate the year over year percent change for homes sold.

```python
homes_sold_years['prev_year_sold'] = homes_sold_years.groupby('state')['homes_sold'].shift(1)
homes_sold_years
```

```python
homes_sold_years['sold_yoy_percent'] = (homes_sold_years['homes_sold'] - homes_sold_years['prev_year_sold']) / homes_sold_years['prev_year_sold'] * 100
homes_sold_years
```

```python
homes_sold_years.dtypes
```

# This is bringing in the top and bottom 5 yoy sale states from SQL

```python
top5_yoy_sales = '''WITH year_ranks AS (SELECT *,
	RANK() OVER(PARTITION BY year ORDER BY sold_yoy_percent DESC)
FROM homes_sold
WHERE sold_yoy_percent IS NOT NULL)
--
SELECT *
FROM year_ranks
WHERE rank BETWEEN 1 AND 5;'''
```

```python
top5_yoy_sales = sql(top5_yoy_sales)
top5_yoy_sales
```

```python
bottom5_yoy_sales = '''WITH year_ranks AS (SELECT *,
	RANK() OVER(PARTITION BY year ORDER BY sold_yoy_percent)
FROM homes_sold
WHERE sold_yoy_percent IS NOT NULL)
--
SELECT *
FROM year_ranks
WHERE rank BETWEEN 1 AND 5;'''
```

```python
bottom5_yoy_sales = sql(bottom5_yoy_sales)
bottom5_yoy_sales
```
